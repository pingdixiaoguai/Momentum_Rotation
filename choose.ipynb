```python
import akshare as ak
import pandas as pd
from datetime import datetime

def get_simple_etf_list():
    try:
        # 从新浪获取ETF列表
        etf_list = ak.fund_etf_category_sina(symbol="ETF基金")
        # etf_list = ak.fund_etf_spot_em()
        
        etf_list = etf_list[['代码', '名称']]
        etf_list = etf_list.rename(columns={'代码': 'symbol', '名称': 'name'}) #  '成交额':'amount'
        
        # etf_list = etf_list.sort_values('amount', ascending=False).reset_index(drop=True)
        etf_list.insert(0, 'rank', range(1, len(etf_list) + 1))
        
        print(f"获取到 {len(etf_list)} 个ETF")
        
        etf_list.to_csv('etf_list.csv', index=False, encoding='utf-8-sig')
        print(f"CSV文件: etf_list.csv")

        print(etf_list.head())
        
        return etf_list
        
    except Exception as e:
        print(f"获取ETF列表失败: {e}")
        return pd.DataFrame()

# 运行简化版本
if __name__ == "__main__":
    get_simple_etf_list()
```
